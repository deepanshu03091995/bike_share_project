```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib.inline
```

#### Loading dataset

```python
file_path = r"D:\GIT\Challenge\bike_share_project\sharing\artifact\data_ingestion\2022-07-13-13-09-50\raw_data\hour.csv"
df = pd.read_csv(file_path)
```

```python
## Check null values
df.isna().sum()
```

```python
# renaming column names to meaningful names 
df = df.rename(columns={
    'weathersit':'weather',
    'yr':'year',
    'mnth':'month',
    'hr':'hour',
    'hum':'humidity',
    'cnt':'count'
})
```

```python
df.columns
```

```python
# change datatypes of columns
cols = ['season', 'year', 'month', 'hour', 'holiday','weekday', 'workingday', 'weather']
for col in cols:
    df[col] = df[col].astype('category')
```

```python
df.info()
```

```python
fig,ax = plt.subplots(figsize=(20,10))
sns.pointplot(data = df,x='hour',y='count', hue ='weekday', ax=ax)
ax.set(title="Count of bikes during weekdays and weekends")
```

### INFERENCES 
1. On weekdays rented bikes were more in morning(6AM - 9 AM) & evening(4PM - 9PM) 
2. On weekends more bike rented in afternoon(10AM- 7PM) 

```python
fig,ax = plt.subplots(figsize=(20,10))
sns.pointplot(data = df,x='hour',y='casual', hue ='weekday', ax=ax)
ax.set(title="Count of bikes during weekdays and weekends : Unregistered users")
```

### INFERENCE
1. Unregistered users uses rented bikes on weekends most

```python
fig,ax = plt.subplots(figsize=(20,10))
sns.pointplot(data = df,x='hour',y='registered', hue ='weekday', ax=ax)
ax.set(title="Count of bikes during weekdays and weekends : Registered users")
```

### INFERENCE
1. Registered users uses rented bikes on weekdays most

```python
fig,ax = plt.subplots(figsize=(20,10))
sns.barplot(data=df,x='month',y='count',ax=ax)
ax.set(title="Count of bikes during different months")
```

### INFERENCE
1. Less bikes were rented in winter season, otherwise it follows normal distribution

```python
corr = df.corr()
plt.figure(figsize=(25,10))
sns.heatmap(corr, annot=True, annot_kws={'size':15})
ax.set(title="NUMERICAL 8 COLS ONLY- [temp,atemp,humidity,windspeed,casual,registered]")
```

### INFERENCES
1. Registered is highly correlated with count
2. Temp and atemp are highly correlated , hence we can delete one amoung them
3. Windspeed is showing very less correlation with dependent column count,hence we can delete Windspeed
4. casual + registered =count, so we can delete casual & registered to avoid multicollinearity

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(20,6))
sns.regplot(x=df['temp'], y=df['count'],ax=ax1)
ax.set(title="Relation between temprature & users")
sns.regplot(x=df['humidity'], y=df['count'],ax=ax2)
ax.set(title="Relation between temprature & users")
```

### INFERENCES

2. With humidity increase users decreases
